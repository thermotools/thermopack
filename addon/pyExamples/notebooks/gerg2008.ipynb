```python
import sys
from thermopack.multiparameter import multiparam
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_context('talk')
sns.set_palette('Dark2')
sns.set_style('ticks')
```

```python
gerg = multiparam("CO2", "GERG2008")
```

## Plot phase envelopes

```python
z = np.array([1.0])
T, P, v = gerg.get_envelope_twophase(5.0e3, z, maximum_pressure=2.0e6, calc_v=True)
Tc = gerg.critical_temperature(1)
vc = gerg.critical_volume(1)
Pc = gerg.critical_pressure(1)
plt.plot(1.0e-3/v, T, "b", label="CO$_2$")
plt.plot([1.0e-3/vc], [Tc], "bo")
#
plt.xlabel(r"$\rho$ (kmol/m$^3$)")
plt.ylabel(r"$T$ (K)")
leg = plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

```python
plt.plot(T, P*1e-6, "b", label=r"CO$_2$")
plt.plot(Tc, Pc*1e-6, "bo")
#
plt.xlabel(r"$T$ (K)")
plt.ylabel(r"$P$ (MPa)")
leg = plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

# Mixtures

```python
gergmix = multiparam("CO2,N2", "GERG2008")
```

## Plot phase envelopes

```python
z = np.array([0.9,0.1])
T, P, v = gergmix.get_envelope_twophase(1.0e5, z, maximum_pressure=2.0e7, calc_v=True)
Tc, vc, Pc = gergmix.critical(z)
plt.plot(1.0e-3/v, T, "b", label="90% CO$_2$ + 10% N$_2$")
plt.plot([1.0e-3/vc], [Tc], "bo")
#
plt.xlabel(r"$\rho$ (kmol/m$^3$)")
plt.ylabel(r"$T$ (K)")
leg = plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

```python
plt.plot(T, P*1e-6, "b", label="90% CO$_2$ + 10% N$_2$")
plt.plot(Tc, Pc*1e-6, "bo")
#
plt.xlabel(r"$T$ (K)")
plt.ylabel(r"$P$ (MPa)")
leg = plt.legend(loc="best")
plt.tight_layout()
plt.show()
```
