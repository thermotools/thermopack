```python
from thermopack.saftvrqmie import saftvrqmie
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_context('talk')
sns.set_palette('Dark2')
sns.set_style('ticks')
```

```python
# Instanciate and init SAFT-VRQ Mie object
qSAFT = saftvrqmie("H2")
qSAFT.set_tmin(temp=2.0)
```

```python
Temperatures = ["20.0", "50.0", "100.0", "300.0"]
r_reduced = np.linspace(0.5, 10, 500)
colors = ["k", "r", "b", "brown", "g", "purple"]
```

```python
for iT, T in enumerate(Temperatures):
    pot = qSAFT.potential(1, 1, r_reduced*qSAFT.sigma[0], float(T))
    plt.plot(r_reduced, pot/qSAFT.eps_div_kb[0], color=colors[iT], label="T="+T+" K")
plt.ylabel(r"$U/\epsilon$")
plt.xlabel(r"$r/\sigma$")
plt.ylim([-1.1, 5.0])
plt.xlim([0.8, 2.5])
leg = plt.legend(loc="best", numpoints=1, ncol=2, frameon=False, fontsize=12)
plt.tight_layout()
plt.savefig("h2_potential.pdf")
```
